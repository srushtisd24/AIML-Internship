```python
#Min Max Normalization
import math
import numpy as np
def min_max(data):
    xi=max(data)
    xj=min(data)
    return [(x - xj) / (xi - xj) for x in data]

numbers = [2, 4, 6, 8, 10]

print("Min-max:",min_max(numbers))
```

```python
#sum
import math
import numpy as np
def std_scalar(data):

    mean=sum(data) / len(data)
    variance=sum((x - mean) ** 2 for x in data) / len(data)
    sd=math.sqrt(variance)
    return [(x - mean) / (sd) for x in data]

numbers = [2, 4, 6, 8, 10]

print("standardscalar:",std_scalar(numbers))
```

```python
#sigmoid function
import math
import numpy as np
numbers = [2, 4, 6, 8, 10]
sigmoid = [1 / (1 + math.exp(-x)) for x in numbers]
print("Sigmoid values:", sigmoid)
```
