```python
data = [10, 20, 30, 40, 50]

mean = sum(data) / len(data)

print("Mean:", mean)
```

```python
def mean(array):
    n = len(array)
    return sum(array) / n
```

```python
mean(np.array([10,20,30,40,50,60]))
```

```python
def variance(array):
    n = len(array)
    m = mean(array)
    return sum((x - m) ** 2 for x in array) / n
```

```python
variance(np.array([1,2,3,4,5,6]))
```
